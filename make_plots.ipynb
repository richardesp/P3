```python
import numpy as np
import matplotlib.pyplot as plt

# set width of bar
barWidth = 0.25
fig = plt.subplots(figsize =(12, 8))

# set height of bar
IT = [12, 30, 1, 8]
ECE = [28, 6, 16, 5]
CSE = [29, 3, 24, 25]

# Set position of bar on X axis
br1 = np.arange(len(mse_tests))
br2 = [x + barWidth for x in br1]

# Make the plot
plt.bar(br1, mse_tests, color ='r', width = barWidth,
    edgecolor ='grey', label ='Tests')
plt.bar(br2, mse_training, color ='g', width = barWidth,
    edgecolor ='grey', label ='Entrenamiento')

# Adding Xticks
plt.title('Comparación del MSE en entrenamiento y tests dado el Nº de neuronas')
plt.xlabel('Ratio de neuronas RBF', fontweight ='bold', fontsize = 15)
plt.ylabel('MSE ', fontweight ='bold', fontsize = 15)
plt.xticks([r + barWidth for r in range(len(mse_tests))],
['0.05', '0.15', '0.25', '0.5'])
plt.legend()
plt.show()
```
