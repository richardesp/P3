```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
seeds = [1, 2, 3, 4, 5]
mse = [74.14, 71.84, 69.54, 74.71, 72.41]
ax.bar(seeds, mse)
ax.set_title('Comparación de resultados de CCR en tests en las distintas semillas')
ax.set_ylabel('CCR')
ax.set_xlabel('Seeds')
plt.show()
```
